```python
#Fungsi Mendasar Matplotlib
#axes.plot(x,y,label='Sesuatu')

x=[0,1,2,3,4,5,6]
y=[10,20,30,40,50,60,70]
fig,axes= plt.subplots(nrows=1,ncols=1,figsize=(12,8))
axes.plot(x,y,label='Percobaan')
```

```python
dates=['2000-01-01','2000-01-02','2000-01-03','2000-01-04','2000-01-05',
       '2000-01-06','2000-01-07','2000-01-08','2000-01-09','2000-01-10']

min_temperature=[20.7,17.9,18.8,14.6,17.4,21.8,20.0,15.8,15.8,15.8]
max_temperature=[34.7,28.9,31.8,28.4,30.8,32.0,25.6,28.8,21.8,22.8]

fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(12,8))

axes.plot(dates,min_temperature,label='Min Temperature')
axes.plot(dates,max_temperature,label='Max Temperature')
axes.legend()
```

Axes.set_xlabel() Mengganti label untuk sumbu x

Axes.set_ylabel() Mengganti label untuk sumbu y

Axes.set_xlim() Membatasi sumbu xnya

Axes.set_ylim() Membatasi sumbu ynya

```python
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(12,8))

axes.plot(dates,min_temperature,label='Min Temperature')
axes.plot(dates,max_temperature,label='Max Temperature')

axes.set_xlabel('Date',fontsize=12)
axes.set_ylabel('Temperature',fontsize=12)

axes.set_title("Minimum and Maximum Temperature",fontsize=12)

axes.set_xticks(dates)
axes.set_xticklabels(dates)
axes.tick_params('x',labelsize=12,labelrotation=45,size=12)

axes.set_ylim(10,40)
axes.set_yticks(np.arange(10,41,2))

axes.legend(fontsize=12,loc='upper left')
```

# DATAFRAMES

Silakan *download*: [Canada.xlsx](./Canada.xlsx)

```python
hai='Canada.xlsx'
df=pd.read_excel(hai,sheet_name='Canada by Citizenship',skiprows=range(20),skipfooter=2)

df.head() #Kurang spesifik ngambil yang mana
```

```python
#Menghapus yang ga mau dipake
df.drop(['AREA','REG','DEV','Type','Coverage'],axis=1,inplace=True)

df.head()
```

```python
df.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
df.head()
```

```python
df.columns=list(map(str,df.columns)) #Mengubah semua nama kolom menjadi string
df.head()
```

```python
df.set_index('Country',inplace=True) #Mengubah Country menajadi index
```

```python
df.head()
```

```python
df['Total']=df.sum(axis=1)
df.head()
```

```python
years=list(map(str,range(1980,2014))) #buatlah sebuah list berupa pemetaan string dimana rangenya itu dari 1980 sampai 2014
years
```

# LINE CHART

```python
algeria = df.loc['Algeria',years]
algeria
```

```python
algeria.index=algeria.index.map(int) #Mengubah tahun menjadi integer
algeria
```

```python
algeria.plot(kind='line',figsize=(10,6))

plt.title('Citizenship Algeria')

plt.ylabel('Banyak Warga')
plt.xlabel('Tahun')
```

```python
albania=df.loc['Albania',years]
albania.index=albania.index.map(int)

albania.plot(kind='line',figsize=(10,6))
algeria.plot(kind='line',figsize=(10,6))

plt.title('Citizenship Algeria')

plt.ylabel('Banyak Warga')
plt.xlabel('Tahun')
```

# SCATTER PLOT

```python
df_total=pd.DataFrame(df[years].sum(axis=0))
df_total
```

```python
df_total.reset_index(inplace=True)
df_total
```

```python
df_total.columns=['year','total']
df_total
```

```python
df_total.plot('year','total',kind='scatter',figsize=(10,6))
```

```python
df_total.plot('year','total',kind='scatter',figsize=(10,6))

plt.title('Scatters Plot')
plt.xticks(rotation='vertical')
```

# PIE CHARTS

```python
df_continents=df.groupby('Continent',axis=0).sum()
df_continents
```

```python
df_continents['Total'].plot(kind='pie',figsize=(5,6),autopct='%1.1f%%') #yang persen gaje nambahin persen
```

```python
df_continents['Total'].plot(kind='pie',figsize=(5,6),autopct='%1.1f%%',startangle=90,shadow=True) #wow diputer 90 derajat plus ada bayangan

plt.axis('equal') #untuk bulat sempurna katanya sih gitu
plt.show()
```

# SUBPLOT

```python
sns.set_style('whitegrid')

fig=plt.figure(figsize=(30,20))
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#pie
df_continents['Total'].plot(kind='pie',ax=ax0) #Maksudnya plotnya ini ditaruh di kanvas ax0
ax0.set_title('Pie Chart')
ax0.axis('equal')

#line
algeria.plot(kind='line',figsize=(15,6),ax=ax1)
albania.plot(kind='line',figsize=(15,6),ax=ax1)
ax1.set_title('Line Plot')

ax1.set_ylabel('Jumlah Warga')
ax1.set_xlabel('Tahun')

plt.show()
```
